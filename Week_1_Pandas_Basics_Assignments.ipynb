```python
df = pd.read_csv("videogamesales.csv")
```

```python
df["global_sales"] = (
    df["NA_Sales"] +
    df["EU_Sales"] +
    df["JP_Sales"] +
    df["Other_Sales"]
)

df_sorted = df.sort_values(by="global_sales", ascending=False)
print(df_sorted)
```

```python
genre_sales = df.groupby("Genre")["global_sales"].sum()

plt.figure(figsize=(8, 5))
genre_sales.plot(kind="bar")
plt.title("Total Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Copies Sold (Millions)")
plt.tight_layout()
plt.show()
```

```python
gta_df = df[df["Name"].str.contains("Grand Theft Auto", case=False, na=False)]

gta_filtered = gta_df[["Name", "Platform", "Year"]].copy()
gta_filtered["EU_JP_Sales"] = gta_df["EU_Sales"] + gta_df["JP_Sales"]

print(gta_filtered)
```

```python
sales_regions = {
    "North America": gta_df["NA_Sales"].sum(),
    "Europe": gta_df["EU_Sales"].sum(),
    "Japan": gta_df["JP_Sales"].sum(),
    "Other": gta_df["Other_Sales"].sum()
}

plt.figure(figsize=(6, 6))
plt.pie(
    sales_regions.values(),
    labels=sales_regions.keys(),
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Global Sales Distribution of Grand Theft Auto Games")
plt.tight_layout()
plt.show()
```
